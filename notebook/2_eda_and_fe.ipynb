# Algerian Forest Fires Dataset

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region 
1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

Add new column with region 

```python
dataset.loc[:122,"Region"] = 0 #Bejaia Region
dataset.loc[122:,"Region"] = 1 #Sidi-Bel Abbes Region Dataset

df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip() #will remove empty spaces 
df.columns
```

```python
df.info()
```

```python
df[['month','day','year','Temperature', 'RH','Ws']] = df[['month','day','year','Temperature', 'RH','Ws']].astype(int)
```

```python
df.info()
```

### Change the other columns to float datatypes

```python
objects = [features for features in df.columns if df[features].dtypes == 'O'] ## 'O' shows the object datatype
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
